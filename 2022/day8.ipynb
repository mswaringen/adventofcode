```python



def vis_check(row_count, col_count,col,row):
    # print(row[col])
    # print(col, width-1)
    # print(row_count, height-1)

    
    left = list(map(int,row[:col]))
    right = list(map(int, row[col+1:]))
    print('horz',row,left,row[col],right)
    
    vert = [row[col] for row in twod]
    vert = "".join(vert)
    
    
    top = list(map(int,vert[:row_count]))
    bottom = list(map(int, vert[row_count+1:]))
    print('vert',vert, top, row[col], bottom)
    
    
    visible = False
    
    
    # check left and right
    print('check left and right')
    max_left = max(left)
    max_right = max(right)
    # print('max_left', max(left))
    # print('max_right', max(right))
    max_top = max(top)
    max_bottom = max(bottom)
    # print('max_top', max_top)
    # print('max_bottom', max_bottom)
        
    if not max_left >= int(row[col]):
        visible = True
        print('Visible from left', max_left, row[col])
    if not max_right >= int(row[col]):
        visible = True
        print('Visible from right', max_right, row[col])

    # check top and bottom
    print('check top and bottom')
    if not max_top >= int(row[col]):
        visible = True
        print('Visible from top', max_top, row[col])
    if not max_bottom >= int(row[col]):
        visible = True
        print('Visible from bottom', max_bottom, row[col])    
    
    return visible






# Part 2


def view_dist(row_count, col, row):
    # print(row[col])
    # print(col, width-1)
    # print(row_count, height-1)
    score_r, score_l, score_t, score_b = 0,0,0,0

    
    left = list(map(int,row[:col]))
    right = list(map(int, row[col+1:]))
    print('horz',row,left,row[col],right)
    
    vert = [row[col] for row in twod]
    vert = "".join(vert)
    
    
    top = list(map(int,vert[:row_count]))
    bottom = list(map(int, vert[row_count+1:]))
    print('vert',vert, top, row[col], bottom)
    
    # check left
    print('check left', left)
    left.reverse()
    
    for i in range(len(left)):
        print(i, left[i], row[col])
        if left[i] >= int(row[col]):
            score_l += 1
            break
        else:
            score_l += 1

    print(score_l)
    
    # check right
    print('check right', right)
    
    for i in range(len(right)):
        print(i, right[i], row[col])
        if right[i] >= int(row[col]):
            score_r += 1
            break
        else:
            score_r += 1
        
    print(score_r)
    
    # check top
    print('check top', top)
    top.reverse()
    
    for i in range(len(top)):
        print(i, top[i], row[col])
        if top[i] >= int(row[col]):
            score_t += 1
            break
        else:
            score_t += 1
        
    print(score_t)

    # check bottom
    print('check bottom', bottom)
    
    for i in range(len(bottom)):
        print(i, bottom[i], row[col])
        if bottom[i] >= int(row[col]):
            score_b += 1
            break
        else:
            score_b += 1
        
    print(score_b)
    
    return score_l * score_r * score_t * score_b


width = len(data[0])
height = len(data)
exterior = (width*2) + ((height-2)*2)

# data = list(map(int, data))
row_count = 0
view_score = 0

for row in data:
    for col in range(width):
        r = view_dist(row_count,col,row)
        print("tree score", r)
        if r > view_score:
            view_score = r
    row_count += 1
```
