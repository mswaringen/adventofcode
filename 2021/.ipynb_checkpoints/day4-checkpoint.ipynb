```python
# extract draw numbers
draw_numbers = list(data[0].split(','))
draw_numbers = [int(i) for i in draw_numbers]
print(draw_numbers)

# remove draw numbers
data = data[2:]
```

```python
# !pip install numpy
import numpy as np
```

```python
def clean_data(data):
    lst = list(data.split(" "))

    while("" in lst):
        lst.remove("")
    for i in range(len(lst)):
        lst[i] = int(lst[i])
    return(lst)

def create_boards(data):
    boards = []
    tmp = []
    for row in range(len(data)):
        if data[row] == '':
            boards.append(tmp)
            tmp = []
        else:
            clean_lst = clean_data(data[row])
            tmp.append(clean_lst)
            if row == len(data)-1:
                boards.append(tmp)
    return boards

def mark_boards(num, boards):
    print("Marking boards...")
    for idx, board in enumerate(boards):
        print("board ", idx)
        print(*board, sep = "\n")
        for row in board:
#             print(num,row)
            if num in row:
                index = row.index(num)
#                 row.pop(index)
                row[index] = 'x'
                print('marked: ',num, row)

def check_boards(boards):
    print("Checking boards:..")
    complete_boards = []
    for board in boards:
        board_t = np.array(board).T.tolist()
        if check_horz(board):
            complete_boards.append(board)
        elif check_horz(board_t):
            complete_boards.append(board)
    if complete_boards:
        return complete_boards
    else:
        return False        
    
def check_horz(board):
    for row in board:
        res = all(ele == 'x' for ele in row)
        if res:
            return True
    return False

def find_score(board, num):
    print("Computing score...")
    unmarked = 0
    for row in board:
        unmarked += sum([x for x in row if isinstance(x,int)])
    final_score = num*unmarked
    print(unmarked, final_score)
    return final_score
                          
def draw_nums(nums, boards):
    for num in nums:
        print("boards left", len(boards))
        print("draw num: ",num)
        mark_boards(num, boards)
        result = check_boards(boards)
        
        if result:
            tmp_board = result[0]
            for board in result:
                tmp_board = board
                b_idx = (boards.index(board))
                boards.pop(b_idx)
                print("Winner, removed board: ", b_idx)
            if not boards:
                return find_score(tmp_board, num)
            else:
                print(find_score(tmp_board, num))
            
```

```python
boards = create_boards(data)
board_col = len(boards[0][0])
print("first board: ",boards[0])
print("board count: ", len(boards))
print("board column count: ", board_col)
```

```python
# boards = [boards[2]]
final = draw_nums(draw_numbers, boards)
```
