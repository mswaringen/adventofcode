```python
test_list = [[0,9],[5,9]]


def find_line(row):
    line = []
    x1, y1, x2, y2 = row[0][0], row[0][1], row[1][0], row[1][1]
    if y1 == y2:
        hi = max(x1,x2)
        lo = min(x1,x2)
        for i in range(lo,hi+1):
            line.append((i, y1))
        print('H ',line)
        return line
    elif x1 == x2:
        hi = max(y1,y2)
        lo = min(y1,y2)
        for i in range(lo,hi+1):
            line.append((x1, i))
        print('V ',line)
        return line
    else:
        return find_diagonal(x1, y1, x2, y2)
        
line = find_line(test_list)
```

```python
lines = []
for row in coors:
    res = find_line(row)
    if res:
        lines.append(res)
```

```python
def find_diagonal(x1, y1, x2, y2):
    
    max_y = max(y1,y2)
    min_y = min(y1,y2)
    max_x = max(x1,x2)
    min_x = min(x1,x2)
    
    if x1 < x2:
        x_chng = 1
    elif x1 > x2:
        x_chng = -1
    else:
        x_chng = 0
        
    if y1 < y2:
        y_chng = 1
    elif y1 > y2:
        y_chng = -1
    else:
        y_chng = 0
    
    x = x1
    y = y1
    line = []
    
#     print(x, x_chng)
#     print(y, y_chng)
    
    for i in range(min_x,max_x+1):
#         print(x,y)
        line.append((x,y))
        x += x_chng
        y += y_chng
    print('D ',line)
    return line
        
```

```python
test_list = [(1,1),(3,3)]
# test_list = [(9,7),(7,9)]

x1 = test_list[0][0]
y1 = test_list[0][1]
x2 = test_list[1][0]
y2 = test_list[1][1]

line = find_diagonal(x1, y1, x2, y2)
```
