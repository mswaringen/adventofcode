```python
from collections import Counter

length = len(data[0])
# print(length)
most_occurence = []
working_lst = data
keep_lst = []

for i in range(length):
    print('bit: ',i)
    
    # Find most common
    tmp_lst = []
    for row in range(len(working_lst)):
        tmp_lst.append(working_lst[row][i])
        
    occur_count = Counter(tmp_lst)
    
    print(occur_count.most_common(1)[0][1], occur_count.most_common(2)[1][1])
    
    
    if occur_count.most_common(1)[0][1] == occur_count.most_common(2)[1][1]:
        most = 1
    else:
        most = Counter(tmp_lst).most_common(1)[0][0]
    most_occurence.append(most)
    print('most_common: ',most)
    
    
    for row in range(len(working_lst)):
        if working_lst[row][i] == most_occurence[i]:
            keep_lst.append(working_lst[row])
    print(keep_lst)
    working_lst = keep_lst
    if len(working_lst) == 1:
        break
    keep_lst = []
    
print(most_occurence)
print(working_lst)


# convert binary to decimal

most_lst = working_lst

# oxy = int(''.join(map(str,most_lst[0])),2)
# e = int(''.join(map(str,epsilon)),2)
# print(g, e, g*e)
print(most_lst)
```

```python
from collections import Counter

length = len(data[0])
most_occurence = []
working_lst = data
keep_lst = []

for i in range(length):
    print('bit: ',i)
    
    # Find least common
    tmp_lst = []
    for row in range(len(working_lst)):
        tmp_lst.append(working_lst[row][i])
        
    occur_count = Counter(tmp_lst)
    print(occur_count.most_common(1)[0][1], occur_count.most_common(2)[1][1])
    
    if occur_count.most_common(1)[0][1] == occur_count.most_common(2)[1][1]:
        most = 0
    else:
        most = Counter(tmp_lst).most_common(2)[1][0]
    most_occurence.append(most)
    print('least_common: ',most)
    
    
    for row in range(len(working_lst)):
        if working_lst[row][i] == most_occurence[i]:
            keep_lst.append(working_lst[row])
    print(keep_lst)
    working_lst = keep_lst
    if len(working_lst) == 1:
        break
    keep_lst = []
    
print(most_occurence)
print(working_lst)


# convert binary to decimal

least_lst = working_lst


print(least_lst)
```
